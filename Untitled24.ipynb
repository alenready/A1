**Handle Outliers in 3 columns**

```python
#  IQR for Age, Fare, Pclass,

# IQR for Age
Q1 = data['Age'].quantile(0.25)
Q3 = data['Age'].quantile(0.75)
IQR_Age = Q3 - Q1

# IQR for Fare
Q1 = data['Fare'].quantile(0.25)
Q3 = data['Fare'].quantile(0.75)
IQR_Fare = Q3 - Q1

# IQR for Pclass
Q1 = data['Pclass'].quantile(0.25)
Q3 = data['Pclass'].quantile(0.75)
IQR_Pclass = Q3 - Q1

# Print the IQR values
print("IQR for Age:", IQR_Age)
print("IQR for Fare:", IQR_Fare)
print("IQR for Pclass:", IQR_Pclass)
```

```python
#  remove outliers from Age,Fare,Pclass

# Remove outliers from Age using IQR
lower_limit_Age = Q1 - 1.5 * IQR_Age
upper_limit_Age = Q3 + 1.5 * IQR_Age

data = data[(data['Age'] >= lower_limit_Age) & (data['Age'] <= upper_limit_Age)]

# Remove outliers from Fare using IQR
lower_limit_Fare = Q1 - 1.5 * IQR_Fare
upper_limit_Fare = Q3 + 1.5 * IQR_Fare

data = data[(data['Fare'] >= lower_limit_Fare) & (data['Fare'] <= upper_limit_Fare)]

# Remove outliers from Pclass using IQR
lower_limit_Pclass = Q1 - 1.5 * IQR_Pclass
upper_limit_Pclass = Q3 + 1.5 * IQR_Pclass

data = data[(data['Pclass'] >= lower_limit_Pclass) & (data['Pclass'] <= upper_limit_Pclass)]

# Print the updated dataframe
data.head()
data.shape
```

```python
# prompt: Outlier  data list of Age, Fare and  Pclass

# Outlier data list of Age, Fare and Pclass
age_outliers = data[(data['Age'] < lower_limit_Age) | (data['Age'] > upper_limit_Age)]['Age'].tolist()
fare_outliers = data[(data['Fare'] < lower_limit_Fare) | (data['Fare'] > upper_limit_Fare)]['Fare'].tolist()
pclass_outliers = data[(data['Pclass'] < lower_limit_Pclass) | (data['Pclass'] > upper_limit_Pclass)]['Pclass'].tolist()

# Print the outlier lists
print("Age outliers:", age_outliers)
print("Fare outliers:", fare_outliers)
print("Pclass outliers:", pclass_outliers)
```

```python
# list the data list of Age, Fare and Pclass after removing outliers

print("List of Age after removing outliers:")
print(data['Age'].tolist())
print("\nList of Fare after removing outliers:")
print(data['Fare'].tolist())
print("\nList of Pclass after removing outliers:")
print(data['Pclass'].tolist())
```
