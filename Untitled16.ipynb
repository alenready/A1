## **Hypothesis testing to clarify any increase in sales after stepping into digital marketing**

```python
# prompt: test hypothesis in sales

# Import the necessary libraries
from scipy.stats import ttest_ind

# Define the null and alternative hypotheses
null_hypothesis = "There is no significant difference in sales before and after the digital ad campaign."
alternative_hypothesis = "There is a significant difference in sales before and after the digital ad campaign."

# Extract the sales data before and after the digital ad campaign
sales_before = df["Sales_before_digital_add(in $)"]
sales_after = df["Sales_After_digital_add(in $)"]

# Perform the independent sample t-test
t_statistic, p_value = ttest_ind(sales_before, sales_after)

# Set the significance level
alpha = 0.05

# Make a decision based on the p-value
if p_value < alpha:
  print(f"Reject the null hypothesis: {alternative_hypothesis}")
else:
  print(f"Fail to reject the null hypothesis: {null_hypothesis}")
```

```python
# prompt: dependency between region and manager

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Read the data from the CSV file
df = pd.read_csv('/content/Sales_add.csv')

# Create a bar chart of the number of sales by region and manager
sns.countplot(x="Region", hue="Manager", data=df)

# Add a title and labels to the chart
plt.title("Number of Sales by Region and Manager")
plt.xlabel("Region")
plt.ylabel("Number of Sales")

# Show the chart
plt.show()

# Calculate the percentage of sales by region and manager
sales_by_region_manager = df.groupby(["Region", "Manager"]).size().unstack()
sales_by_region_manager_pct = sales_by_region_manager.apply(lambda x: 100 * x / x.sum(), axis=1)

# Print the percentage of sales by region and manager
print(sales_by_region_manager_pct)
```

```python


import pandas as pd
from scipy import stats

# Read the data from the CSV file
df = pd.read_csv('/content/Sales_add.csv')

# Create a contingency table of the number of sales by region and manager
contingency_table = pd.crosstab(df['Region'], df['Manager'])

# Perform a chi-square test for independence between region and manager
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

# Print the results of the chi-square test
print("Chi-square statistic:", chi2)
print("P-value:", p)
print("Degrees of freedom:", dof)
print("Expected frequencies:")
print(expected)

# Interpret the results of the chi-square test
if p < 0.05:
  print("There is a statistically significant association between region and manager.")
else:
  print("There is no statistically significant association between region and manager.")
```

```python
# Hypothesis chi-square test for independence between region and manager
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

# Set the significance level
alpha = 0.05

# Make a decision based on the p-value
print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}"'\n')
if p < alpha:
  print(f"Reject the null hypothesis: There is a statistically significant association between region and manager.")
else:
  print(f"Fail to reject the null hypothesis: There is no statistically significant association between region and manager.")
```
