```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# 1: Consider the following Python dictionary data and Python list labels:
data = {'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
        'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
        'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
import pandas as pd
import numpy as np
df = pd.DataFrame(data, index=labels)
df
```

```python
# prompt: **11.** Calculate the sum of all visits in `df` (i.e. the total number of visits).

total_visits = df['visits'].sum()
print(f"Total visits: {total_visits}")
```

```python
# prompt: **12.** Calculate the mean age for each different animal in `df`.

df.groupby('animal')['age'].mean()
```

```python
# prompt: **13.** Append a new row 'k' to `df` with your choice of values for each column. Then delete that row to return the original DataFrame.

# Create a dictionary with the desired values for the new row
new_row = {'animal': 'owl', 'age': 5, 'visits': 4, 'priority': 'yes'}

# Append the new row to the DataFrame
df.loc['k'] = new_row

# Print the DataFrame with the new row
print(df)

# Delete the row 'k' from the DataFrame
df.drop('k', inplace=True)
print('\n', df)
```

```python
# prompt: **14.** Count the number of each type of animal in `df`.

df['animal'].value_counts()
```

```python
# prompt: **15.** Sort `df` first by the values in the 'age' in *decending* order, then by the value in the 'visits' column in *ascending* order (so row `i` should be first, and row `d` should be last).

df.sort_values(by=['age', 'visits'], ascending=[False, True])
```
