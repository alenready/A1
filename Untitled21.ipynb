# ***IQR METHOD HANDLING OUTLIERS on beer_servings and total_litres_of_pure_alcohol ***:

```python
# prompt: iqr handling on beer_servings

import numpy as np
q1 = df['beer_servings'].quantile(0.25)
q3 = df['beer_servings'].quantile(0.75)
print('q1=',q1,'q3=',q3)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
print('lower_bound=', lower_bound, 'upper_bound=', upper_bound)
df['beer_servings'] = np.where((df['beer_servings'] < lower_bound) | (df['beer_servings'] > upper_bound), df['beer_servings'].median(), df['beer_servings'])
```

```python
# prompt: outliers for beer_servings

q1 = df['beer_servings'].quantile(0.25)
q3 = df['beer_servings'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

outliers = df[((df['beer_servings'] < lower_bound) | (df['beer_servings'] > upper_bound))]

print(f"Number of outliers for beer_servings: {outliers.shape[0]}")
```

```python
# prompt: dataframe showing with out outliers

df_without_outliers = df[~((df['beer_servings'] < lower_bound) | (df['beer_servings'] > upper_bound))]
df_without_outliers
```

```python
# prompt: compare the difference on box plot on beer_servings with and with out ouliers

import matplotlib.pyplot as plt
import seaborn as sns

# With outliers
plt.figure(figsize=(10, 6))
sns.boxplot(df['beer_servings'])
plt.title('Boxplot of Beer Servings with Outliers')
plt.show()

# Without outliers
plt.figure(figsize=(10, 6))
sns.boxplot(df_without_outliers['beer_servings'])
plt.title('Boxplot of Beer Servings without Outliers')
plt.show()
```

# ***z-score handling outliers***

```python
# prompt: z score handling on beer_servings

import pandas as pd
import numpy as np

# Calculate the z-score for each data point in the 'beer_servings' column
z_scores = np.abs((df['beer_servings'] - df['beer_servings'].mean()) / df['beer_servings'].std())

# Define a threshold for identifying outliers
threshold = 3

# Select rows where the z-score is greater than the threshold
outliers = df[z_scores > threshold]

# Print the outlier rows
print("Outlier rows:")
print(outliers)

# Remove the outlier rows from the DataFrame
df_without_outliers = df[z_scores <= threshold]

# Print the DataFrame without outliers
print("\nDataFrame without outliers:")
print(df_without_outliers)
```

```python
# prompt: zscore on total_litres_of_pure_alcohol

zscore = (df['total_litres_of_pure_alcohol'] - df['total_litres_of_pure_alcohol'].mean()) / df['total_litres_of_pure_alcohol'].std()

df['zscore'] = zscore

df.head()
```
