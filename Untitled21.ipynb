# ***IQR METHOD HANDLING OUTLIERS***

```python
Q1=df.total_litres_of_pure_alcohol.quantile(0.25)
Q3=df.total_litres_of_pure_alcohol.quantile(.75)
Q1,Q3
```

```python
IQR=Q3-Q1
IQR
```

```python
lower_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR
lower_limit,upper_limit
```

```python
# prompt: boxplot

import matplotlib.pyplot as plt
import seaborn as sns
sns.boxplot(df['total_litres_of_pure_alcohol'])
plt.show()
```

```python
# prompt: plot outliers

import matplotlib.pyplot as plt

plt.figure()
plt.scatter(df['total_litres_of_pure_alcohol'], df['beer_servings'])
plt.xlabel('total_litres_of_pure_alcohol')
plt.ylabel('beer_servings')

# Plot the outlier points
plt.scatter(df[df['total_litres_of_pure_alcohol'] < lower_limit]['total_litres_of_pure_alcohol'], df[df['total_litres_of_pure_alcohol'] < lower_limit]['beer_servings'], c='red')
plt.scatter(df[df['total_litres_of_pure_alcohol'] > upper_limit]['total_litres_of_pure_alcohol'], df[df['total_litres_of_pure_alcohol'] > upper_limit]['beer_servings'], c='red')

plt.show()
```

```python
# prompt: plot outliers in total_litres_of_pure_alcohol

import matplotlib.pyplot as plt

# Create a boxplot of the total_litres_of_pure_alcohol column
plt.boxplot(df['total_litres_of_pure_alcohol'])

# Add a title and labels to the plot
plt.title('Boxplot of Total Litres of Pure Alcohol')
plt.xlabel('Total Litres of Pure Alcohol')
plt.ylabel('Value')

# Show the plot
plt.show()
```

```python
# prompt: Identify the outlier rows using the lower and upper limits.

outlier_rows = df[(df['total_litres_of_pure_alcohol'] < lower_limit) | (df['total_litres_of_pure_alcohol'] > upper_limit)]
print(outlier_rows)
```

```python
# prompt: Remove the outlier rows from the DataFrame.

df_without_outliers = df[(df['total_litres_of_pure_alcohol'] >= lower_limit) & (df['total_litres_of_pure_alcohol'] <= upper_limit)]

print(df_without_outliers)
```

```python
# prompt: Print the outlier rows and the DataFrame without outliers.

# Print the outlier rows
print("Outlier rows:")
print(outlier_rows)

# Print the DataFrame without outliers
print("\nDataFrame without outliers:")
print(df_without_outliers)
```

# ***z-score handling outliers***

```python
# prompt: zscore on df

zscore = (df['total_litres_of_pure_alcohol'] - df['total_litres_of_pure_alcohol'].mean()) / df['total_litres_of_pure_alcohol'].std()

df['zscore'] = zscore

df.head()
```
