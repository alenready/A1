```python
# prompt: program in python without database

# List of customers
customers = [
    {"id": 1, "name": "John", "password": "password123", "balance": 25000},
    {"id": 2, "name": "Jane", "password": "password456", "balance": 50000},
    {"id": 3, "name": "Bob", "password": "password789", "balance": 100000},
]

# Main menu
def main_menu():
    while True:
        print("\nCustomer Menu:")
        print("1. Customer Login")
        print("2. New Customer Sign in")
        print("3. Exit")
        choice = input("Enter option: ")
        if choice == '1':
            customer_login()
        elif choice == '2':
            new_customer_sign_in()
        elif choice == '3':
            print("Exiting...")
            break
        else:
            print("Invalid option, please try again.")

# Customer login
def customer_login():
    cust_name = input("Enter Customer Name: ")
    password = input("Enter Password: ")
    for customer in customers:
        if customer["name"] == cust_name and customer["password"] == password:
            print(f"Welcome {cust_name}...")
            account_operations(customer["id"])
            return
    print("Invalid credentials.")

# New customer sign in
def new_customer_sign_in():
    cust_name = input("Enter Customer Name: ")
    password = input("Enter Password: ")
    new_customer = {"id": len(customers) + 1, "name": cust_name, "password": password, "balance": 0}
    customers.append(new_customer)
    print(f"Customer {cust_name} created successfully.")

# Account operations
def account_operations(customer_id):
    while True:
        print("\nAccount Menu:")
        print("a. Amount Deposit")
        print("b. Amount Withdrawal")
        print("c. Check Balance")
        print("d. Exit")
        choice = input("Enter option: ")
        if choice == 'a':
            amount_deposit(customer_id)
        elif choice == 'b':
            amount_withdrawal(customer_id)
        elif choice == 'c':
            check_balance(customer_id)
        elif choice == 'd':
            break
        else:
            print("Invalid option, please try again.")

# Amount deposit
def amount_deposit(customer_id):
    while True:
        try:
            amount = float(input("Enter amount to deposit: "))
            if amount > 0:
                for customer in customers:
                    if customer["id"] == customer_id:
                        customer["balance"] += amount
                        check_balance(customer_id)
                        break
                break
            else:
                print("Amount must be greater than 0.")
        except ValueError:
            print("Invalid amount, please enter a number.")

# Amount withdrawal
def amount_withdrawal(customer_id):
    while True:
        try:

            amount = float(input("Enter amount to withdraw: "))
            if amount > 0  :
                for customer in customers:
                    if customer["id"] == customer_id:
                        customer["balance"] -= amount
                        check_balance(customer_id)
                        break
                break
            else:
                print("Invalid amount, please enter a valid amount.")
        except ValueError:
            print("Invalid amount, please enter a number.")

# Check balance
def check_balance(customer_id):
    for customer in customers:
        if customer["id"] == customer_id:
            balance = customer["balance"]
            break
    print(f"Current balance: {balance}")

# Main program
if __name__ == "__main__":
    main_menu()
```
